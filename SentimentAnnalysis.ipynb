```python
# import libraries
import pandas as pd
import numpy as np
import scipy.stats as st
import os
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import csv

# used to analyse data

from nltk.sentiment.vader import SentimentIntensityAnalyzer
#a pre-trained model for sentiment analysis

from nltk.corpus import stopwords
#contains a list of common words that are usually removed from text during processing.

from nltk.tokenize import word_tokenize
#is used to split text into individual words.

from nltk.stem import WordNetLemmatizer
#is used to convert words to their base form.




# Load the amazon review dataset (need to fix this) 

df = pd.read_csv('https://raw.githubusercontent.com/pycaret/pycaret/master/datasets/amazon.csv', on_bad_lines='skip')

df
```

```python
# punctuaction remover from De-Bugging
import string
def remove_punctuation(text):
    # Create a translation table that maps each punctuation character to None
    translator = str.maketrans('', '', string.punctuation)
    # Use translate method to remove punctuation characters
    cleaned_text = text.translate(translator)
    return cleaned_text

#number romover from De-Bugging
import re

def remove_numbers(input_string):
    # Use a regular expression to replace all digits with an empty string
    return re.sub(r'\d', '', input_string)

# create preprocess_text function
def preprocess_text(text):

    # Tokenize the text

    tokens = word_tokenize(str(text).lower())




    # Remove stop words

    filtered_tokens = [token for token in tokens if token not in stopwords.words('english')]




    # Lemmatize the tokens

    lemmatizer = WordNetLemmatizer()

    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]
    
    

    # Join the tokens back into a string

    processed_text = ' '.join(lemmatized_tokens)
    
    

    # Remove Punctuation 

    #next_text = remove_punctuation(processed_text)
    
    
    
    # Remove Numbers
    
    #final_text = remove_numbers(next_text)
    
    
    
    return lemmatized_tokens
        
def token(nums):
    # need to make these into strings maybe???
    tokens = word_tokenize(str(nums).lower())
    return tokens

# apply the function df

df['Positive'] = df['Positive'].apply(token)

df['reviewText'] = df['reviewText'].apply(preprocess_text)
    
df
#each colomb in a dataframe is a series
```

```python
from sklearn.model_selection  import train_test_split
#splits the data into a training and testing set
train,test = train_test_split(dn,random_state=42,test_size=0.3)
```

```python
#creating the independent and dependent variables

X_train = [' '.join(words) for (words, category) in train]
X_test = [' '.join(words) for (words, category) in test]
y_train = [category for (words, category) in train]
y_test = [category for (words, category) in test]
#for the independent variables it also joins the tokens back into strings so that they can be re-tokenized by the vecorization process
```

```python
#text vectorization and BOW model
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

#vectorize
tfidf_vec = TfidfVectorizer(min_df = 10, token_pattern = r'[a-zA-Z]+')
#ignores words that are used less than 10 times (gets rid of rare non-informative words)
# r'[a-zA-Z]+' is the token patern 
    #This pattern will match words composed of only letters and ignore any other characters like numbers or punctuation.
#make bag of words
X_train_bow = tfidf_vec.fit_transform(X_train) # fit train
X_test_bow = tfidf_vec.transform(X_test) # transform test
#This will generate BOW matrix X_train_bow and X_test_bow based on your input data, using the vocabulary and settings specified in tfidf_vec.
    #print(e)
```

```python
#Train Model
from sklearn.linear_model import LogisticRegression

model_lg = LogisticRegression()
#good for true/false (2 variable) data
model_lg.fit(X_train_bow, y_train)
```

```python
print(model_lg.score(X_test_bow, y_test))
```
